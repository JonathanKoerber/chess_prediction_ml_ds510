```python
from chess_env_bitboard import BitBoardEnv
from dqn_agent import DQN_Agent
import numpy as np
import time
import random
```

```python
env = BitBoardEnv()
agent = DQN_Agent(env, state_shape=(12, 8, 8), num_actions=64)
```

```python
num_episodes = 50
for episode in range(num_episodes):
    state = env.reset()
    done = env.done
    episode_experiences = []  # Accumulate experiences during the episode

    while not done:
        action = agent.choose_action(state)
        next_state, reward, done = env.step(action)
        episode_experiences.append((state, action, reward, next_state, done))
        state = next_state

    # Store experiences in the replay buffer
    for experience in episode_experiences:
        agent.store_experience(*experience)

    # Sample and train on a batch of experiences
    batch_size = 32  # Adjust the batch size as needed
    if len(agent.replay_buffer) >= batch_size:
        batch = np.random.choice(agent.replay_buffer, size=batch_size, replace=False)
        states, actions, rewards, next_states, dones = zip(*batch)
        agent.train(states, actions, rewards, next_states, dones)

    print(f"Episode {episode+1}/{num_episodes} done")
```
